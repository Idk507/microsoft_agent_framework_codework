```python
import os 
from dotenv import load_dotenv
from agent_framework import ChatAgent
from agent_framework.azure import AzureAIAgentClient
from azure.identity.aio import AzureCliCredential
from agent_framework.devui import serve 
import asyncio 

```

```python
from agent_framework_azure_ai._chat_client import AzureAIAgentClient
async def create_docs_agent(project_endpoint : str, model : str) -> ChatAgent :
    chat_client = AzureAIAgentClient(
        async_credential=AzureCliCredential(),
        project_endpoint=project_endpoint,
        model_deployment_name=model
    )
    try :
        agent = chat_client.create_agent(
        name = "Docs Agent",
        instructions= "Use the provided documentation to answer user queries as accurately as possible. If the information is not available in the documentation, respond with 'I don't know'.",
        )
        print("Docs Agent created successfully.")
        return agent
    except Exception as e :
        print("Error creating Docs Agent:", str(e))
        return None

```

```python
async def load_agent() :
        load_dotenv()
        project_endpoint = os.getenv("AZURE_AI_PROJECT_ENDPOINT")
        model = os.getenv("AZURE_AI_DEPLOYMENT_MODEL")
        agent = await create_docs_agent(project_endpoint, model)
        return agent

```

```python
agent = asyncio.run(load_agent())
serve(entities=[agent], auto_open= True, tracing_enabled=True)
```
